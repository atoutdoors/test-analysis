```python
import pandas as pd
import matplotlib.pyplot as plt
import os 
pd.set_option('display.max_rows', None)
```

```python
#Input parameters - change these
SNs = ["1489"]
SN = 1489
test = 'valve'
date = "20230517" #latest date will include all data from test
root = "G:\Shared drives\WHOI Argo\BGC\Float Tracking\pressure_test_data\\" + test + '\\' + str(SN)
```

```python
#Open and merge data
data = pd.DataFrame()
for sn in SNs:
    crt = pd.read_pickle(root+"/F"+sn+"_"+date+".pkl")
    crt["SN"] = sn
    data = pd.concat([data, crt])
```

```python
data['HUMIDITY'] = data['HUMIDITY'].astype(float)
```
